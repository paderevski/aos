```python
import random
import collections
import pandas
```

```python
CLASS_SIZE = 10
```

```python
def sample(num_collisions):
    birthdays = random.choices(range(31), k=CLASS_SIZE)
    counts = collections.Counter(birthdays)
    return num_collisions in counts.values()
```

```python
def repeat_sample(num_times, num_collisions):
    return [sample(num_collisions) for x in range(num_times)].count(True)/num_times
```

```python
data = [repeat_sample(1000,i) for i in range(7)]
```

```python
df = pandas.DataFrame(data, columns=["Probability"])
```

```python
df.plot.bar()
```
